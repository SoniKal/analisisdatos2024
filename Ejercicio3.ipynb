```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("movies_updated.csv")

# Gráfico 1: Distribución de géneros (genre) con un gráfico de barras
plt.figure(figsize=(10, 6))
sns.countplot(x='genre', data=df, palette='pastel')
plt.title('Distribución de géneros de películas')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

# Gráfico 2: Relación entre presupuesto (budget) y ganancias (gross) con un scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='budget', y='gross', data=df, alpha=0.7)
plt.title('Relación entre presupuesto y ganancias')
plt.xlabel('Presupuesto (en millones)')
plt.ylabel('Ganancias (en millones)')
plt.show()

# Gráfico 3: Distribución de calificaciones (rating) con un pie chart
rating_counts = df['rating'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribución de calificaciones de películas')
plt.show()

# Análisis de los resultados
# 1. En el primer gráfico, se muestra la distribución de géneros de películas. Se observa que el género Comedia es el más común, seguido de drama y Acción.

# 2. En el segundo gráfico, se representa la relación entre el presupuesto y las ganancias de las películas. Se puede apreciar una poca dispersión de los datos.

# 3. En el tercer gráfico, se presenta la distribución de calificaciones de películas. La mayoría de las películas tienen calificaciones de PG-13 y R, mientras
#    que las calificaciones G son menos comunes. Esto puede indicar una tendencia hacia películas dirigidas a audiencias más maduras.

# 4. Estos gráficos nos proporcionan una visión diversa de diferentes aspectos de las películas, incluyendo género, relación entre presupuesto y ganancias,
#    y distribución de calificaciones. Nos ayudan a comprender mejor la composición y las tendencias dentro del conjunto de datos.
```
