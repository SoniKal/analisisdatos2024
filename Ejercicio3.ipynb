```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv("movies_updated.csv")

# Grfico 1: Distribucin de gneros (genre) con un grfico de barras
plt.figure(figsize=(10, 6))
sns.countplot(x='genre', data=df, palette='pastel')
plt.title('Distribucin de gneros de pelculas')
plt.xlabel('Gnero')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

# Grfico 2: Relacin entre presupuesto (budget) y ganancias (gross) con un scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='budget', y='gross', data=df, alpha=0.7)
plt.title('Relacin entre presupuesto y ganancias')
plt.xlabel('Presupuesto (en millones)')
plt.ylabel('Ganancias (en millones)')
plt.show()

# Grfico 3: Distribucin de calificaciones (rating) con un pie chart
rating_counts = df['rating'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribucin de calificaciones de pelculas')
plt.show()

# Anlisis de los resultados
# 1. En el primer grfico, se muestra la distribucin de gneros de pelculas. Se observa que el gnero Comedia es el ms comn, seguido de drama y Accin.

# 2. En el segundo grfico, se representa la relacin entre el presupuesto y las ganancias de las pelculas. Se puede apreciar una poca dispersin de los datos.

# 3. En el tercer grfico, se presenta la distribucin de calificaciones de pelculas. La mayora de las pelculas tienen calificaciones de PG-13 y R, mientras
#    que las calificaciones G son menos comunes. Esto puede indicar una tendencia hacia pelculas dirigidas a audiencias ms maduras.

# 4. Estos grficos nos proporcionan una visin diversa de diferentes aspectos de las pelculas, incluyendo gnero, relacin entre presupuesto y ganancias,
#    y distribucin de calificaciones. Nos ayudan a comprender mejor la composicin y las tendencias dentro del conjunto de datos.
```
